The important part: Portfolio returns. We use the portfolios created by make_portfolio.ipynb to calculate portfolio returns for each decile.

```python
import numpy as np
from tqdm import tqdm
import warnings
from pandas.errors import SettingWithCopyWarning
warnings.simplefilter(action='ignore', category=(SettingWithCopyWarning))

ticker_df = pd.read_csv("data\\ticker-return-daily.csv")
strat_df = pd.read_pickle("data\portfolio\\neg_portfolio_enhanced-bow.pkl")

def cal_portfolio_return(strat_df_path,ticker_df_path="data\\ticker-return-daily.csv"):
    ticker_df = pd.read_csv(ticker_df_path)
    strat_df = pd.read_pickle(strat_df_path)
    strat_df = pd.DataFrame(strat_df)
    strat_df = strat_df.T
    strat_df.reset_index(inplace=True)
    strat_df.rename(columns={'index': 'date'}, inplace=True)
    
    if 'score' in strat_df.columns:
        strat_df.drop(columns=['score'], inplace=True)
        
    strat_df.rename(columns={'ticker':'tic'}, inplace=True)
    strat_df = strat_df.explode(column=['tic', 'weight','decile'])
    ticker_df_table = ticker_df.pivot_table(index='tic', columns='date', values='ret')
    dt = strat_df['date'].unique()
    return_series =pd.Series([])
    
    # Iterate over the deciles
    for decile in tqdm(range(1,11)):
        decile_return_series =pd.Series([])
        ith_decile = strat_df[strat_df['decile'] == decile]
        ith_decile.drop(columns=['decile'], inplace=True)
        
        # Calculate decile portfolio return
        for i in range(len(dt)-1):
            start_dt = dt[i]
            end_dt = dt[i+1]
            weight_df = ith_decile[(ith_decile['date'] == start_dt)]
            if not decile_return_series.empty:
                cur_value = decile_return_series.iloc[-1]
                weight_df.loc[:,'weight'] = weight_df['weight'] * cur_value # rebalance the portfolio
            tickers = ith_decile[(ith_decile['date'] == start_dt)]['tic']
            sub_df = ticker_df.loc[(ticker_df['date'] >= start_dt) & (ticker_df['date'] < end_dt) & (ticker_df['tic'].isin(tickers))]
            sub_dt = sub_df['date'].unique()
            #print(sub_dt)
            sub_df = ticker_df_table[sub_dt]
            sub_df = sub_df[sub_df.index.isin(tickers)]
            sub_df = sub_df.apply(lambda x: 1+x)
            sub_df = sub_df.fillna(1).reset_index() # Fix the NaN bug first
            #return_df = pd.DataFrame(sub_df.prod(axis=1),columns=['return'])
            merged_df = pd.merge(sub_df, weight_df, how='left', on= ['tic']).drop(columns=['date'])
            #print(merged_df.columns)
            cols = merged_df.columns[:-1]
            cols = cols.insert(1, merged_df.columns[-1])
            merged_df = merged_df[cols].set_index('tic')
            decile_return_series = pd.concat([decile_return_series,merged_df.cumprod(axis=1).drop(columns=['weight']).sum(axis=0)])
        decile_return_series['decile'] = decile
        return_series = pd.concat([return_series,decile_return_series])
    return return_series
#return_series =  cal_portfolio_return("data\portfolio\\neg_portfolio_enhanced-bow.pkl","data\\ticker-return-daily.csv")
```
